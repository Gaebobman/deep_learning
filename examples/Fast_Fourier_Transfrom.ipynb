```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
# Create a signal
time = np.linspace(0, 2 * np.pi, 1000)
signal = np.sin(10 * time) + np.cos(20 * time)
```

```python
plt.plot(time, signal)
plt.xlabel('Time (s)')
plt.ylabel('Signal')
```

```python
# Compute the Fast Fourier Transform (FFT) of the Signal
fft = np.fft.fft(signal)
```

```python
# Compute the Power spectrum  of the Signal
power_spectrum = np.square(np.abs(fft))
```

```python
# Compute the frequency axis
n = signal.size
# freq_axis = np.fft.fftfreq(n)
freq_axis = np.fft.fftfreq(n, 2*np.pi/n)
```

```python
plt.plot(freq_axis, power_spectrum)
plt.xlabel('Frequency (Hz)')
plt.ylabel('Power')
```
